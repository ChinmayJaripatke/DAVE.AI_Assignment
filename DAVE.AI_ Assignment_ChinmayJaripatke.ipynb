```python
import pandas as pd
```

# Importing the customer one dataset:

```python
data1 = pd.read_csv(r"D:\Chinmay\Dave_AI Assignment\customer_data_one.csv")
```

```python
data1
```

# Importing the customer two dataset:

```python
data2 = pd.read_csv(r"D:\Chinmay\Dave_AI Assignment\customer_data_two.csv")
```

```python
data2
```

# Merging the datasets using outer join:

```python
data = pd.merge(data1, data2, how = "outer")
```

```python
data
```

# Keeping the records from the second dataset only (if any duplicates):

```python
data_nondup = data.drop_duplicates(subset = ['first_name'], keep = 'last').reset_index(drop = True)
```

```python
data_nondup
```

# Sorting the dataset by the first name of the customer:

```python
data_sort = data_nondup.sort_values(by = 'first_name')
```

```python
data_sort
```

# Setting the Mobile Number as the index:

```python
data_sort.set_index('mobile_number', inplace = True)
```

```python
data_sort
```

# Converting the dataframe to JSON format:

```python
json_data = data_sort.to_json(orient = 'index')
```
